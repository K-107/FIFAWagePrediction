# 12. 20년 데이터

```python
df20 = pd.read_csv('./dataset/players_20.csv')
df20
```

```python
df20 = df20[['sofifa_id','short_name','age','nationality','club','overall','potential',
             'wage_eur','value_eur','contract_valid_until']]


df20 = df20.rename(columns={'sofifa_id':'ID','short_name':'Name','age':'Age20',
                            'nationality':'Nationality','club':'Club20',
                            'overall':'Overall20','potential':'Potential20',
                            'wage_eur':'Wage20','value_eur':'Value20',
                            'contract_valid_until':'ContractValidUntil20'})

df20
```

## 전처리

```python
# ContractValidUntil
df20.ContractValidUntil20.unique()
```

```python
df20.ContractValidUntil20.isnull().sum()
```

```python
df20.ContractValidUntil20.replace(np.nan, '0', inplace=True)
df20.drop(df20[df20['ContractValidUntil20']==0].index, inplace=True)
df20.ContractValidUntil20 = df20.ContractValidUntil20.astype(int)
df20.ContractValidUntil20
```

```python
df20.ContractValidUntil20.isnull().sum()
```

```python
# Value20 원화로 바꿈.
df20.Value20 = df20.Value20*1300/10000
df20.Value20
```

```python
df20.Value20.unique()
```

```python
# Value20이 0인 선수 제거
df20.drop(df20[df20['Value20']==0].index, inplace=True)
```

```python
# Wage20 원화로 바꿈.
df20.Wage20 = df20.Wage20*1300/10000
df20.Wage20 = df20.Wage20.astype(int)
df20.Wage20
```

```python
df20.Wage20.unique()
```

```python
df20['LogWage20'] = np.log1p(df20.Wage20)*10
```

```python
df = pd.read_pickle('./dataset/나혼자FIFA_이적예상WAGE.pkl')
df19 = df[['ID','Age','Club','Overall','Potential','Wage','Value','ContractValidUntil']]

df19 = df19.rename(columns={'Age':'Age19','Club':'Club19','Overall': 'Overall19',
                            'Potential':'Potential19','Wage':'Wage19',
                            'Value':'Value19','ContractValidUntil':'ContractValidUntil19'})

df19
```

```python
df19.info()
```

```python
df19.ContractValidUntil19.unique()
```

```python
def change_date(date):
    years = ['2018','2019','2020','2021','2022','2023','2024','2025','2026']
    if date in years:
        out = date
    else:
        out = date.split(',')[1]
    return int(out) 
    
    
df19.ContractValidUntil19 = df19.ContractValidUntil19.apply(lambda x: change_date(x))
df19.ContractValidUntil19
```

# 이름만 바뀐 팀 19년도 팀명으로 바꿔줌.

```python
def change_team_name(team):
    out = ''
    if team == 'Guangzhou R&F FC':
        out = 'Guangzhou R&F; FC'
        
    elif team == 'CSA - AL':
        out = 'América FC (Minas Gerais)'
        
    elif team == 'Paraná':
        out = 'Fortaleza'
        
    elif team == 'Vitória':
        out = 'Goiás'
        
    elif team == 'Sport Club do Recife':
        out = 'Avaí FC'       

    else:
        out = team

    return out

df1920.Club20 = df1920.Club20.apply(lambda x: change_team_name(x))
df1920.Club20
```
